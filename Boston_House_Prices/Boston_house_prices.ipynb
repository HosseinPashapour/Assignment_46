## Import Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from perceptron import Perceptron
from matplotlib.animation import FuncAnimation
```

## Load Boston dataset from the scikit-learn library

### Preprocessing on dataset

```python
data = pd.read_csv('Boston.csv')
data = data.drop('Unnamed: 0', axis=1)
data.rename(columns={'medv': 'price'}, inplace=True)
```

```python
data.corr()
```

## This dataset has 13 features for each house. According to data correlation and your analysis, choose 2 features for X choose 1 target for Y: price

```python
X = np.array([data['rm'], data['zn']])
X = X.T
Y = np.array(data[['price']])
```

## Split the dataset to train and test

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=.2)
```

## Solve the regression problem on 2 dimensions (2 features) using our perceptron algorithm

```python
model = Perceptron(lr_w=.001, lr_b=.01, epochs=30)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

## Show results on a 3D plot and draw the y_pred as a plane on it as animation

```python
model.w
```

```python
X1 = x_train[:,0]
X2 = x_train[:,1]
x1, x2 = np.meshgrid(X1, X2)
eq = model.weights[110][0][0] * x1 + model.weights[110][0][1] * x2 + model.b[0][0]
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')

ax.clear()

ax.scatter(X1 , X2 , y_train,  color='red',alpha=0.5)
ax.plot_surface(x1, x2, eq)
ax.set_xlabel('rm')
ax.set_ylabel('zn')
ax.set_zlabel('price')
def rotate(angle):
    ax.view_init(azim=angle)

rot_animation = FuncAnimation(fig, rotate, frames=np.arange(0, 362, 2), interval=100)
rot_animation.save('boston_y_pred.gif', dpi=80, writer='imagemagick')


plt.show()
```
